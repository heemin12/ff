```python
# 극좌표계 설명과 예제

import numpy as np
import matplotlib.pyplot as plt

## 극좌표계란? 
'''
극좌표계는 평면 위의 점의 위치를 각도와 거리로 표현하는 좌표계입니다. 
- r: 원점에서부터의 거리
- θ: x축에서부터 반시계 방향으로 잰 각도 (단위: radian)

직교좌표계 (x, y)와의 관계식:
x = r * cos(θ)  
y = r * sin(θ)
r = sqrt(x^2 + y^2)
θ = arctan(y/x)
'''
```

def km_per_hour_to_meters_per_second(speed_kmh):
    # 1 km/h는 0.277778 m/s와 같음
    return speed_kmh * 0.277778

```python
## 예제 1: 직교좌표계로 표현된 점을 극좌표계로 변환하기  
x = 3
y = 4
r = np.sqrt(x**2 + y**2) 
theta = np.arctan2(y, x)  # y, x 순서 주의 (-pi ~ pi 사이의 값)
print(f"직교좌표계 (x, y) = ({x}, {y})")
print(f"극좌표계 (r, θ) = ({r:.2f}, {theta:.2f}) rad ({np.degrees(theta):.1f}°)")
```

```python
## 예제 2: 극좌표계로 표현된 점을 직교좌표계로 변환하기
r = 2
theta = np.pi/3  # 60도
x = r * np.cos(theta)
y = r * np.sin(theta)
print(f"극좌표계 (r, θ) = ({r}, {theta:.2f}) rad")
print(f"직교좌표계 (x, y) = ({x:.2f}, {y:.2f})")
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='b', width=0.005, headwidth=5)
ax.text(x, y+0.1, f'(r, θ) = ({r}, {theta:.2f})', fontsize=12, color='b', ha='center')
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.set_title('극좌표계 → 직교좌표계', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
## 예제 3: 직교좌표계로 표현된 점을 극좌표계로 변환하기  
x = -8
y = -5
r = np.sqrt(x**2 + y**2) 
theta = np.arctan2(y, x)  # y, x 순서 주의 (-pi ~ pi 사이의 값)
print(f"직교좌표계 (x, y) = ({x}, {y})")
print(f"극좌표계 (r, θ) = ({r:.2f}, {theta:.2f}) rad ({np.degrees(theta):.1f}°)")
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(5, 5))
ax.plot([theta], [r], 'ro')
ax.quiver(0, 0, theta, r, angles='xy', scale_units='xy', scale=1, color='r', width=0.01, headwidth=5)
ax.text(theta, r, f'(x, y) = ({x}, {y})', fontsize=12, color='r', ha='center', va='bottom')
ax.set_rticks([2, 4, 6, 8, 10])
ax.set_rlabel_position(-22.5)
ax.grid(True)
ax.set_title('직교좌표계 → 극좌표계', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 벡터란?
'''
벡터는 크기와 방향을 가진 물리량입니다. 
벡터는 숫자(스칼라)와 달리, 방향성을 가지고 있어 더 많은 정보를 담고 있습니다.

예를 들어, 속도는 빠르기(크기)와 함께 방향도 가지고 있기 때문에 벡터입니다.
반면, 질량은 크기만 있고 방향은 없기 때문에 스칼라입니다.  

벡터는 화살표로 표현하며, 시작점에서 끝점으로 향하는 방향을 가집니다.
'''
```

```python
v = np.array([3, 1])  # 벡터 v
w = np.array([2, 4])  # 벡터 w

# 벡터 v와 w 그리기
fig, ax = plt.subplots(figsize=(5, 5))

# 벡터 v
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(v[0], v[1], 'v', fontsize=14, color='b')

# 벡터 w
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.text(w[0], w[1], 'w', fontsize=14, color='r')

ax.set_xlim(-1, 6)
ax.set_ylim(-1, 6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.set_title('평면에서의 벡터', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 예제 4: 단위벡터 x^와 y^ 그리기
fig, ax = plt.subplots(figsize=(5, 5))

ax.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='r', width=0.008, headwidth=6)
ax.quiver(0, 0, 0, 1, angles='xy', scale_units='xy', scale=1, color='g', width=0.008, headwidth=6)
ax.text(1.1, 0.1, '$\hat{x}$', fontsize=16, color='r')
ax.text(0.1, 1.1, '$\hat{y}$', fontsize=16, color='g')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.grid(True)
ax.set_title('직각좌표계의 단위벡터', fontsize=16)
ax.set_aspect('equal')

plt.tight_layout()
plt.show()
```
