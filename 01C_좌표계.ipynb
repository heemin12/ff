```python
# 극좌표계 설명과 예제

import numpy as np
import matplotlib.pyplot as plt

## 극좌표계란? 
'''
극좌표계는 평면 위의 점의 위치를 각도와 거리로 표현하는 좌표계입니다. 
- r: 원점에서부터의 거리
- θ: x축에서부터 반시계 방향으로 잰 각도 (단위: radian)

직교좌표계 (x, y)와의 관계식:
x = r * cos(θ)  
y = r * sin(θ)
r = sqrt(x^2 + y^2)
θ = arctan(y/x)
'''
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# 가능한 font list 확인
import matplotlib.font_manager as fm
f = [f.name for f in fm.fontManager.ttflist]
print(f)

# 확인 이후
plt.rc('font', family='Malgun Gothic')
mpl.rcParams['axes.unicode_minus'] = False
```

```python
## 예제 1: 직교좌표계로 표현된 점을 극좌표계로 변환하기 - 슬라이드 9p
x = 3
y = 4
r = np.sqrt(x**2 + y**2) 
theta = np.arctan2(y, x)  # y, x 순서 주의 (-pi ~ pi 사이의 값)
print(f"직교좌표계 (x, y) = ({x}, {y})")
print(f"극좌표계 (r, θ) = ({r:.2f}, {theta:.2f}) rad ({np.degrees(theta):.1f}°)")
```

```python
## 예제 2: 극좌표계로 표현된 점을 직교좌표계로 변환하기 - 슬라이드 10p
r = 2
theta = np.pi/3  # 60도
x = r * np.cos(theta) # theta의 단위는 라디안
y = r * np.sin(theta) # theta의 단위는 라디안
print(f"극좌표계 (r, θ) = ({r}, {theta:.2f}) rad")
print(f"직교좌표계 (x, y) = ({x:.2f}, {y:.2f})")
```

```python
import numpy as np

def polar_to_cartesian_rounded(r, theta_degrees, decimal=2):
    # 각도를 라디안으로 변환
    theta_radians = np.radians(theta_degrees)
    
    # 극좌표계에서 직교좌표계로 변환
    x = r * np.cos(theta_radians)
    y = r * np.sin(theta_radians)
    
    # 반올림하여 반환
    x_rounded = round(x, decimal)
    y_rounded = round(y, decimal)
    
    return x_rounded, y_rounded

# 문제에서 제시된 극좌표계 좌표
r = 15.0 # 반지름, 미터
theta_degrees = 120 # 각도, 도

# 극좌표계에서 직교좌표계로 변환하여 결과 출력
x_rounded, y_rounded = polar_to_cartesian_rounded(r, theta_degrees)
print(f"직교좌표계 (x, y) = ({x_rounded}, {y_rounded})")
```

```python
import numpy as np

def cartesian_to_polar_rounded(x, y, decimal=2):
    # 반지름 r 계산
    r = np.sqrt(x**2 + y**2)
    
    # 각도 theta 계산
    theta = np.arctan2(y, x)
    
    # 각도를 0~360도 범위로 변환
    theta_degrees = np.degrees(theta) % 360
    
    # 반올림하여 반환
    r_rounded = round(r, decimal)
    theta_rounded = round(theta_degrees, decimal)
    
    return r_rounded, theta_rounded

# 문제에서 제시된 직교좌표계 좌표
x = 3.0
y = 4.0

# 직교좌표계에서 극좌표계로 변환하여 결과 출력
r_rounded, theta_rounded = cartesian_to_polar_rounded(x, y)
print(f"극좌표계 (r, θ) = ({r_rounded}, {theta_rounded})")
```

```python
r = 15
theta = np.pi/2  # 60도
x = r * np.cos(theta) # theta의 단위는 라디안
y = r * np.sin(theta) # theta의 단위는 라디안
print(f"극좌표계 (r, θ) = ({r}, {theta:.2f}) rad")
print(f"직교좌표계 (x, y) = ({x:.2f}, {y:.2f})")
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='b', width=0.005, headwidth=5)
ax.text(x, y+0.1, f'(r, θ) = ({r}, {theta:.2f})', fontsize=12, color='b', ha='center')
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.set_title('극좌표계 → 직교좌표계', fontsize=14)
plt.tight_layout()
plt.show()
```
